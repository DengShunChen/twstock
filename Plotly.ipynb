```python
from stocktools import StockTools
import chart_studio.plotly as py
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
st = StockTools('2018-03-01','2018-08-31')
```

```python
#st.strdate = '2019-09-01'
#st.enddate = '2019-09-31'
#st.save_stock('2801')
```

```python
st.strdate = '2018-03-01'
st.enddate = '2019-12-31'
st.read_stock('2801')
```

```python
st._stock_anal('2801')
```

```python
ma_pd = st.ma_pd.reset_index()
st_pd = st.stock_pd.reset_index()
norstd = st.norstd.reset_index()
```

```python
#fig = go.Figure()
fig = make_subplots(specs=[[{"secondary_y": True}]])

fig.add_trace(go.Scatter(x=st_pd.date, y=st_pd.close, name="收盤價",
                         line_color='black'))
fig.add_trace(go.Scatter(x=ma_pd.date, y=ma_pd.ma03, name="3日均價",
                         line_color='lime'))
fig.add_trace(go.Scatter(x=ma_pd.date, y=ma_pd.ma05, name="5日均價",
                         line_color='royalblue'))
fig.add_trace(go.Scatter(x=ma_pd.date, y=ma_pd.ma08, name="8日均價",
                         line_color='darkviolet'))
fig.add_trace(go.Scatter(x=ma_pd.date, y=ma_pd.ma20, name="20日均價",
                         line_color='red'))
fig.add_trace(go.Scatter(x=ma_pd.date, y=ma_pd.ma60, name="60日均價",
                         line_color='orange'))
fig.add_trace(go.Scatter(x=norstd.date, y=norstd[0], name="變異係數",
                         line_color='lightblue'),secondary_y=True,)

fig.update_layout(title_text='%s %s'%(st.id,st.name),
                  xaxis_rangeslider_visible=True)

fig.update_yaxes(title_text="<b>價格</b>", secondary_y=False)
fig.update_yaxes(title_text="<b>變異係數</b>", secondary_y=True)

fig.update_layout(legend_orientation="h",legend=dict(x=0, y=-0.5))
fig.update_layout(
    xaxis=go.layout.XAxis(
        rangeselector=dict(
            buttons=list([
                dict(count=1,
                     label="1m",
                     step="month",
                     stepmode="backward"),
                dict(count=6,
                     label="6m",
                     step="month",
                     stepmode="backward"),
                dict(count=1,
                     label="YTD",
                     step="year",
                     stepmode="todate"),
                dict(count=1,
                     label="1y",
                     step="year",
                     stepmode="backward"),
                dict(step="all")
            ])
        ),
        rangeslider=dict(
            visible=True
        ),
        type="date"
    )
)


fig.show()
#py.iplot(fig, filename='%s %s'%(st.id,st.name))
```
